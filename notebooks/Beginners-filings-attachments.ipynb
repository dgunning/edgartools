# Beginner's Guide: filings vs. filing and attachments of a filing

This is a beginner's guide to clarify the concept of filings vs. filing and to introduce the use of attachments in edgartools

```python
#!pip install -U edgartools   # Uncomment this to install the package in ter


from edgar import *
from edgar.xbrl import *

set_identity("John Doe john.doe@myEmail.com")   # Tell the SEC who you are
```

```python
# Company("NVDA") with cik = 1045810

# Get a list of filings meeting the conditions on cik, form, and filing_date
filings = (Company(1045810).get_filings(form=["10-K", "10-Q"])   # get either 10-K or 10-Q
            .filter(filing_date="2023-01-01:2023-12-31")
          )
filings          
```

```python
# Filtering for "10-K" alone returns an object of the EntityFilings class (note the plural in EntityFilings)
filings_still = filings.filter(form="10-K")

print(type(filings_still)) # show the type of the object
print(dir(filings_still))  # show the methods of the object
filings_still
```

```python
# filings_still as an object of the EntityFilings class has no attachments method 

filings_still.attachments
```

# filings_still[0] vs. filings_still: Note the different types and methods  

```python

# Choose the only filing in the list filings_still
filing = filings_still[0]  

# Note the different types and methods for filing = filings_still[0], as opposed to the list filings_still
print(type(filing))  # note EntityFiling, as opposed to EntityFilings
print(dir(filing))   # EntityFiling has the attachments method
```

```python
# open the filing and its home page in a browser
filing.open()  
filing.open_homepage()  
```

```python
# Show the list of attachments in a table

filing.attachments
```

```python
# View attachments[19]

# 19  â”‚ R2.htm                         â”‚ Audit Information                                            â”‚ ğŸŒ  HTML  
filing.attachments[19].view()   
```

```python
# Show the methods of the attachments object
# Note: Scroll to the right to see all the methods

print(dir(filing.attachments)) 
```

```python
# More details on using the query method of attachments 

help(filing.attachments.query)
```

```python

# Select attachments[19] by querying on document 
# (i.e., the Document column in the previously displayed attachments table)

# 19  â”‚ R2.htm                         â”‚ Audit Information                                            â”‚ ğŸŒ  HTML  
print(filing.attachments.query("document == 'R2.htm'"))

#View the selected attachment
filing.attachments.query("document == 'R2.htm'").data_files[0].view()
```
