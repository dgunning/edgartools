## Install edgartools

## Import edgar and set your identity

```python
from edgar import *

set_identity("demo@funds.com")
```

## Get a fund by its ticker

```python
fund_class = get_fund("KINCX")
fund_class
```

```python
get_fund("S000005031")
```

### Fund

```python
fund = fund_class.fund
fund
```

### Series

```python
fund.get_series()
```

## Get the fund's filings

```python
fund.get_filings()
```

## Filtering Fund filings

```python
npx_filings = fund.get_filings(form="N-PX")
npx_filings
```
