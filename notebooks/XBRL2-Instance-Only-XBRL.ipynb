# Parse Instance-Only XBRL Documents from SEC Filings with Python -- Free, No API Key

Use **edgartools** to parse instance-only XBRL documents (like 424B2 prospectuses) that contain XBRL data without full presentation linkbases -- completely free, no API key required.

# Instance Only XBRL

```python
from edgar import *
from edgar.xbrl import *
```

```python
filings = get_filings(index="xbrl").tail(40000)
filings
```

```python
filing = filings[0]
filing
```

```python
xbrl = XBRL.from_filing(filing)
xbrl
```
